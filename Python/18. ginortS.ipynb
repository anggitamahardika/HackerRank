Source Problem: [ginortS](https://www.hackerrank.com/challenges/ginorts/problem?isFullScreen=true)

```python
word = input()

lowercase, uppercase, odd, even = "", "", "", ""

for i in word:
    if (i.islower()):
        lowercase += i
        lowercase = ''.join(sorted(lowercase))
    elif(i.isupper()):
        uppercase += i
        uppercase = ''.join(sorted(uppercase))
    elif(i.isdigit() and int(i)%2 != 0):
        even += i
        even = ''.join(sorted(even))
    else:
        odd += i
        odd = ''.join(sorted(odd))
        

print(lowercase+uppercase+even+odd)
```
