Source Problem: [Nested Lists](https://www.hackerrank.com/challenges/nested-list/problem?isFullScreen=true)

```python
if __name__ == '__main__':
    data = []
    nilai = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        data.append([name, score])
        nilai.append(score)
        
    nilai = list(dict.fromkeys(nilai))
    nilai.sort()
    data.sort(key=lambda x:x[0])
    data.sort(key=lambda x:x[1])

    for i in data:
        if i[1] == nilai[1]:
            print(i[0])
```
