Source Problem: [Alphabet Rangoli](https://www.hackerrank.com/challenges/alphabet-rangoli/problem?isFullScreen=true)

```python
def print_rangoli(size):
    alphabet='abcdefghijklmnopqrstuvwxyz'
    
    if(size == 1):
        print('a')
        return
    
    strip=(size*2)-1
    
    for i in range(1, size+1):
        # print "-" prefix
        count_strip = (strip - 2*i)
        print(count_strip*'-', end='')
        
        # print top left side
        for j in alphabet[size-1:size-1-i:-1]:
            print('-' + j, end="")
        
        # print center
        if i == size :
            for j in alphabet[size-1:size-i:-1]:
                if j == alphabet[size-1]:
                    print(j, end="")
                else: 
                    print('-' + j, end="")
            
            print('-a', end='')
            for j in alphabet[1:size]:
                print('-' + j, end="")
        
        # print top right side
        if i >= 2 and i < size:            
            for j in alphabet[size+1-i:size]:
                print('-' + j, end="")
                 
        # print "-" suffix
        print((count_strip+1)*'-', end='')
            
        print('')
        
    for i in reversed(range(1, size)):
        # print "-" prefix
        count_strip = (strip - 2*i)
        print(count_strip*'-', end='')
        
        # print bottom left side
        for j in alphabet[size-1:size-1-i:-1]:
            print('-' + j, end="")
        
        if i >= 2 and i < size:            
            for j in alphabet[size+1-i:size]:
                print('-' + j, end="")
        print((count_strip+1)*'-', end='')
        print('')

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
