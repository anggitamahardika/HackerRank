Source Problem: [Designer Door Mat](https://www.hackerrank.com/challenges/designer-door-mat/problem?isFullScreen=true)

```python
arr = list(map(int, input().split()))

# TOP
for i in range(arr[0]//2):
    count_pattern = 2*(i+1) - 1
    strip = (arr[1] - (count_pattern*3))//2
    print('-' * strip + '.|.' * count_pattern + '-' * strip)

# CENTER
strip = (arr[1] - 7)//2
print('-' * strip + 'WELCOME' + '-' * strip)  

# BOTTOM
for i in reversed(range(arr[0]//2)):
    count_pattern = 2*(i+1) - 1
    strip = (arr[1] - (count_pattern*3))//2
    print('-' * strip + '.|.' * count_pattern + '-' * strip)
```
